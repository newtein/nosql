```python
colors=['b','c','y','g','k','m','C2','C2','C3','C4']
for family in nosql:
    plt.close()
    plt.subplots(figsize=(5,2.5),dpi=300)
    count=0
    t=0
    for year in data_o['mongodb']:
        #print(year,[data_o[dbs][year] for dbs in nosql[family]])
        plt.bar([i+t+2 for i in range(len(nosql[family]))],[data_o[dbs][year] for dbs in nosql[family]],color=colors,width=3)
        count+=1
        t+=15
    for color,dbs in zip(colors,nosql[family]):
        plt.scatter(-5,-5,label=dbs,color=color)
    plt.xlim(0,None)
    plt.ylim(0,None)
    plt.xticks([i+len(nosql[family])/2 for i in range(0,t,15)],[i for i in data_o[dbs]])
    plt.legend(loc="upper left")
    #plt.title(family.upper())
    print(family)
    plt.ylabel("Number of questions asked",size=10)
    plt.show()
```
