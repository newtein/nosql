```python
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import numpy as np
sns.set()
```

```python
data=pickle.load(open("ts/Scoreresults.txt","rb"))
data_o=pickle.load(open("ts/Scoreresults-2.txt","rb"))
```

```python
print(data_o.keys())

nosql={}
nosql["keyvalue"]=['redis','dbm','dynamo','aerospike', 'memcachedb', 'riak', 'berkeley db']
nosql["column"]=['cassandra','hbase','accumulo','druid','vertica']
nosql["graph"]=['neo4j','orientdb','allegrograph', 'virtuoso']
nosql["document"]=['mongodb','couchbase','apache couchdb','arangodb', 'basex','cosmos db', 'marklogic', 'rethinkdb']
nosql["Hadoop"]=['hadoop','hadoop 2', 'hadoop3','hdfs']
nosql["NewSQL"]=['voltdb', 'â€žcockroachdb', 'nuodb', 'hana', 'memsql']
nosql["Comparison"]=["redis","cassandra","neo4j","mongodb","hdfs","memsql"]
nosql["MongoDB"]=['python', 'php', 'java', 'js', '.net', 'ruby', 'node', 'c++', 'c#']
nosql["Cassandra-Versions"]=['c2.0', 'c3.0', 'c0.7', 'c2.1', 'c2.2']
nosql["Cassandra"]=['spark', 'c-jdbc', '.net library', 'python driver', 'node driver', 'c# driver']
```

```python
import math

for i in data_o:
    sum=[]
    for year in data_o[i]:
        #print(i,year,data_o[i][year])
        sum.extend(data_o[i][year])
    try:
        #print(sum)
        print(i,np.mean(sum))
    except:
        print(i,0,"Error")
```

```python
colors=['b','y','g','c','k','m','C1','C2','C3','C4']
for family in nosql:
    plt.close()
    plt.subplots(figsize=(9,4),dpi=100)
    count=0
    t=0
    for year in data_o['mongodb']:
        #print(year,[data_o[dbs][year] for dbs in nosql[family]])
        plt.bar([i+t for i in range(len(nosql[family]))],[np.mean(data_o[dbs][year]) for dbs in nosql[family]],color=colors)
        count+=1
        t+=13
        
    for color,dbs in zip(colors,nosql[family]):
        plt.scatter(-1,-1,label=dbs,color=color)
    plt.xlim(-2,None)
    plt.ylim(0,None)
    print(len(nosql[family])/2)
    plt.xticks([i+len(nosql[family])/2 for i in range(0,t,13)],[i for i in data_o[dbs]])
    plt.legend(loc="upper right")
    plt.title(family.upper())
    plt.ylabel("Mean score on each question")
    plt.show()
```
